```python
# Import openCV libaries
import cv2
```

```python
# Read image file
img = cv2.imread("../input/images-2/Annotation_Birds.png")
# Show images
cv2.imshow("Annotation_Birds", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Create a viewImage funcation for resizing
def viewImage(img, name_of_window):
    cv2.nameWindow(name_of_window, cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    scalling = 80
    
    # Changing width and height
    width = int(img.shape[1] * scalling / 100)
    height = int(img.shape[0] * scalling / 100)
    dim = (width, height)
    resized = cv2.resize(img,dim,interpolation = cv2.INTER_AREA)
    
    viewImage(resized, "Resized Image")
    
```
